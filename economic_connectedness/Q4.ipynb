According to the paper and the assignment instructions, due to its definition, we need to take the mean exposure to high-parental-SES individuals by high school and divide it by two.

Then , as we can see, both x and y axis are percentages. So, we need to multiply them by 100.

```python
df1.exposure_parent_ses_hs = df1.exposure_parent_ses_hs /2
```

```python
df1.bias_parent_ses_hs = df1.bias_parent_ses_hs*100
```

```python
df1.exposure_parent_ses_hs = df1.exposure_parent_ses_hs*100
```

We wil use both the df1 and the hs dataframes together in the plot.

However, we need to import ggplot from plotnine.

We will also use adjustText to annotate the selected high schools.

The plot looks like this:

```python
from plotnine import *
import adjustText

adjust_text_dict = {
    'expand_points': (2, 1),
    'arrowprops': {
        'arrowstyle': '->',
        'color': '#7ee4c2',
        'lw': 1.5
    }
}

ggplot(df1) +\
    aes(x='exposure_parent_ses_hs', y='bias_parent_ses_hs') +\
    geom_point(alpha = 0.1, na_rm=True) +\
    geom_point(data = hs,
              color = '#83e3e3',
              size = 4,
              alpha = 0.9) +\
    labs(x='Share of high-parental-SES students (%)', y='Friending bias among low-parental-SES students (%)') +\
    scale_x_continuous(limits = (0,100),breaks = (0,20,40,60,80)) +\
    scale_y_reverse(limits = (-15,30), breaks= (30,20,10,0,-10)) +\
    geom_label(mapping = aes(label='high_school_name'),
                adjust_text = adjust_text_dict,
                alpha = 0.9, data = hs) +\
    theme(figure_size=(8,6))
    
```
