As we can see, both datasets contain columns for the percentage of the races.

```python
zips = pd.read_csv('data/social_capital_zip.csv')
```

We have to do the same for the colleges. 

So, we read the 'social_capital_college.csv'.

We will merge the clg and the slg dataframes on the college id (column named college).

```python
sclg = pd.read_csv('data\social_capital_college.csv')
```

Let's begin with the neighborhoods.

We keep the columns that we need; the shares of the races along with the population, zip and the number of children with below-national-median parental household income.

```python
data = ['zip','num_below_p50','pop2018','share_white_2018','share_black_2018','share_natam_2018','share_asian_2018','share_hawaii_2018','share_hispanic_2018']
```

```python
zip_covariates = zip_covariates[data]
```

```python
zip_covariates
```

According to the paper and the assignment instructions, the Herfindahlâ€“Hirschman index for our case is calculated as follows:

We take the sum of the squares of the share of each race and subtract it from 1.

```python
zip_covariates['hhi'] = 1 - (zip_covariates['share_white_2018'].pow(2) +
                             zip_covariates['share_black_2018'].pow(2) +
                             zip_covariates['share_natam_2018'].pow(2) +
                             zip_covariates['share_asian_2018'].pow(2) +
                             zip_covariates['share_hawaii_2018'].pow(2) +
                             zip_covariates['share_hispanic_2018'].pow(2))
```

Our dataframe looks like this:

```python
zip_covariates
```

It's time to merge, as said before, the two dataframes on the zip code.

```python
zip_merged = pd.merge(zips, zip_covariates, on='zip', how='inner')
```

```python
zip_merged
```

We only keep what we will use; The zip code, the mean of the neighborhood friending bias, the hhi and the number of children with below-national-median parental household income

```python
 zip_merged = zip_merged[['zip','nbhd_bias_zip','hhi','num_below_p50_x']]
```

We need to multiply the mean of the neighborhood friending bias according to the replication package instructions.

```python
zip_merged['nbhd_bias_zip'] = zip_merged['nbhd_bias_zip']*100
```

```python
zip_merged
```

```python
zip_merged.rename(columns={'nbhd_bias_zip':'bias'},inplace = True)
```

It is crucial to remove all the NaN values from the 'num_below_p50_x' , so that the bins are created as shown in the papers/diagram.

```python
zip_merged = zip_merged.dropna(subset=['num_below_p50_x'])
```

```python
zip_merged
```

It is time to create the 20 equal bins, based on the hhi.

We will use qcut, since we want 20 equal-sized buckets.

```python
bins = pd.qcut(zip_merged['hhi'],20)
```

```python
bins
```

```python
zip_merged['bins'] = bins
```

```python
# group = zip_merged.groupby('bins')['hhi','bias','num_below_p50_x'].mean()
```

Now, let's create the weights for the weighted mean for the number of children with below-national-median parental household income.

Each bin contains some Herfindahl-Hirschman Indexes that are in the bin range.

However, we need to take into account that the hhi of each zip in the bin should have differnet weight, because of the population and the number of children with below-national-median parental household income of each zip.

That is what we do in the function 'wm'.

Then, we need to group the data based on the bins created.

This time, we take the mean of the bias(as stated in the instructions for the y axis) and we actually applythe function 'wm' on the hhi of each bin.



```python
import numpy as np

wm = lambda x: np.average(x, weights=zip_merged.loc[x.index, "num_below_p50_x"])

group1 = zip_merged.groupby("bins")['hhi','bias','num_below_p50_x'].agg(bias=("bias", "mean"),  
                                                      hhi_weighted_mean=("hhi", wm))
```

The group1 variable now contains all the information needed for the plotting of the 20 bins.

```python
group1
```

To be sure, we plot the group1 dataframe as a scatterplot.

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_style('dark')
sns.set(rc={'figure.figsize':(7, 5)})

_ = sns.regplot(x='hhi_weighted_mean',
                y ='bias',
                color='#fd8820',
                data = group1,
#                 ci =  0
                
#                 palette = my_pal,
#                 ax = ax,
#                 facet_kws={'sharey': False, 'sharex': False}
               )


_.set(xlabel="Racial Diversity (Herfindahl-Hirschman Index) in Group",
     ylabel = "Friending Bias among Low-SES Individuals (%)")
```

Now, we will repeat the steps for the colleges.


According to the paper and the assignment instructions, the Herfindahlâ€“Hirschman index for our case is calculated as follows:

We take the sum of the squares of the share of each race and subtract it from 1.

But first, we need to find the share of the white people in each college.

```python
clg['white_share_fall_2000'] = 1 -(clg['asian_or_pacific_share_fall_2000'] + 
                                  clg['black_share_fall_2000'] +
                                  clg['hisp_share_fall_2000'] )
```

```python
clg['hhi'] = 1 -(clg['asian_or_pacific_share_fall_2000'].pow(2) + 
                clg['black_share_fall_2000'].pow(2) +
                 clg['hisp_share_fall_2000'].pow(2) +
                clg['white_share_fall_2000'].pow(2))
```

```python
clg
```

Now, it is time to merge the two dataframes about the colleges to combine the mean number of students per cohort with the hhis.

```python
clg_merged = pd.merge(clg,sclg, on='college', how = 'inner')
```

```python
clg_merged
```

We keep only the columns that we need.

```python
data_clg  = ['college',
             'bias_own_ses_college',
             'hhi',
             'mean_students_per_cohort_x'
            ]
```

```python
clg_merged = clg_merged[data_clg]
```

The merged dataframe looks like this:

```python
clg_merged
```

We rename the bias_own_ses_college for the sake of simplicity and multiply it by 100.

```python
clg_merged.rename(columns={'bias_own_ses_college':'bias'},inplace = True)
```

```python
clg_merged['bias'] = clg_merged['bias'] * 100
```

Again, it is crucial to get rid of all the NaN values. We focus on the mean number of students per cohort, since this is the column of the weights.

```python
clg_merged = clg_merged.dropna(subset=['mean_students_per_cohort_x'])
```

```python
clg_merged
```

Again, we create the 20 bins we want using qcut.

```python
bins_clg = pd.qcut(clg_merged['hhi'],20)
```

```python
bins_clg
```

```python
clg_merged['bins'] = bins_clg
```

```python
clg_merged
```

The procedure is the same, but with different columns. 

We create a function that calculates the weights of the different hhis based on the mean_students_per_cohort of each college in the bin.

We then create a new dataframe, group2, which groups the merged dataframe by the bins we have created.

```python
import numpy as np

wm = lambda x: np.average(x, weights=clg_merged.loc[x.index, "mean_students_per_cohort_x"])

group2 = clg_merged.groupby("bins")['hhi','bias','mean_students_per_cohort_x'].agg(bias=("bias", "mean"),  
                                                      hhi_weighted_mean=("hhi", wm))
```

Now, the second dataframe about the colleges is ready to be plotted.

However, since we want two scatterplots in the same figure, we will combine the two dataframes(group1 and group2) into one.

Be careful, it is not a merge, since they do not share something in commmon, but a concatenation.

We need to insert a category column, which will help differentiate the two types of data we have.

```python
group2['categ'] = 'College'
group1['categ'] = 'Neighborhood'
```

The 2 dataframes look like this:

```python
group1
```

```python
group2
```

Finally, it is time to concatenate them and start plotting.

```python
df = pd.concat([group1,group2])
```

```python
df
```

We will use seaborn's lmplot, which is quite helpful for plotting scatterpolts along with their regression lines.

The plot looks like this:

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_style('dark')
sns.set(rc={'figure.figsize':(14, 12)})

# fig, ax = plt.subplots()

_ = sns.lmplot(x='hhi_weighted_mean',
                y ='bias',
                hue = 'categ',
                data = df,
                hue_order=['College', 'Neighborhood'],
                palette= ['#2078b4','#fd8820'],
                markers=['D','o'],
                ci = 95,
                height= 4.5,
                aspect =1.5
               )

sns.move_legend(_, loc="upper left", bbox_to_anchor=(0.115, 0.96), title=None,frameon=True)

_.set(xlabel="Racial Diversity (Herfindahl-Hirschman Index) in Group",
     ylabel = "Friending Bias among Low-SES Individuals (%)")
```
